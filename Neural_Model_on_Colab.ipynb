```
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
from google.colab import files
symptom_file = files.upload()
train_file = files.upload()
test_file = files.upload()
```

```
def create_cases_dict(symptoms, cases):
  cases_dict = {}
  for symptom in symptoms:
    symptom = symptom.strip().lower()
    cases_dict[symptom] = [0] * len(cases)

  for index,case_symptoms in enumerate(cases):
    case_symptoms = case_symptoms.split(';')
    for symptom in case_symptoms:
      symptom = symptom.strip().lower()
      if symptom != '':
        cases_dict[symptom][index] = 1

  return cases_dict
```

```
def get_symptoms_data(data):
  all_dict = {}
  for values in data:
    values = values.split(';')
    for symptom in values:
      symptom = symptom.strip().lower()
      if symptom in all_dict:
        all_dict[symptom] = all_dict[symptom] + 1
      else:
        all_dict[symptom] = 1
  return all_dict
```

```
import io
symptoms_list_data = pd.read_excel(io.BytesIO(symptom_file['list2.xlsx']))
train_cases_data = pd.read_excel(io.BytesIO(train_file['train.xlsx']))
test_cases_data = pd.read_excel(io.BytesIO(test_file['test.xlsx']))

all_data = get_symptoms_data(train_cases_data['SIGNS AND SYMPTOMS'])
relevant_data = get_symptoms_data(train_cases_data['RELEVANT SIGNS AND SYMPTOMS']) 

train_cases_all_symptoms = create_cases_dict(symptoms_list_data['SIGNS AND SYMPTOMS'], train_cases_data['SIGNS AND SYMPTOMS'])
train_cases_relevant_symptoms = create_cases_dict(symptoms_list_data['SIGNS AND SYMPTOMS'], train_cases_data['RELEVANT SIGNS AND SYMPTOMS'])

test_cases_all_symptoms = create_cases_dict(symptoms_list_data['SIGNS AND SYMPTOMS'], test_cases_data['SIGNS AND SYMPTOMS'])
test_cases_relevant_symptoms = create_cases_dict(symptoms_list_data['SIGNS AND SYMPTOMS'], test_cases_data['RELEVANT SIGNS AND SYMPTOMS'])

train_all_symptoms_df = pd.DataFrame(train_cases_all_symptoms)
train_relevant_symptoms_df = pd.DataFrame(train_cases_relevant_symptoms)

test_all_symptoms_df = pd.DataFrame(test_cases_all_symptoms)
test_relevant_symptoms_df = pd.DataFrame(test_cases_relevant_symptoms)

print(train_relevant_symptoms_df)
```

```
from keras.models import Sequential
from keras.layers import Dense
```

```
model = Sequential()
model.add(Dense(53, activation="relu", input_shape=(53,)))
model.add(Dense(53, activation="sigmoid"))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(train_all_symptoms_df, train_relevant_symptoms_df, epochs=4, batch_size=56, verbose=1)
```

```
model.evaluate(test_all_symptoms_df, test_relevant_symptoms_df, batch_size=12)

predictions = model.predict(test_all_symptoms_df[:1])
print(predictions)
```
